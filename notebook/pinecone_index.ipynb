```python
from pinecone import Pinecone,ServerlessSpec
import os
from dotenv import load_dotenv
load_dotenv()
api_key = os.environ.get("pinecone_api_key")
pinecone = Pinecone(api_key=api_key)
```

Creating storage index

```python
# index agent would use to store and retrieve vector embeddings
index = pinecone.create_index(
    name = "stock-analyst-storage-index",
    dimension = 1536,
    spec = ServerlessSpec(
        cloud = "aws",
        region = "us-east-1",
    )
    )
```

```python
pinecone_index = pinecone.Index("stock-analyst-storage-index")
```

```python
print("Index connected:", pinecone_index)
```
